```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Import Data1.csv file to python.
df = pd.read_csv (r'C:\Dash Toh\AGH university\Cyber-Physical System 2nd-cycle degree\1st semester\Data Analytics\Laboratory\Lab1\Data1.csv')
```

```python
#Set first column as the index.
df.set_index('Unnamed: 0')
```

```python
#Plot all columns as time series.
f, ax = plt.subplots(3,2, figsize=(15,5))
df.plot('Unnamed: 0','theta_1', ax=ax[0][0], title = 'Theta_1')
df.plot('Unnamed: 0','theta_2', ax=ax[0][1], title = 'Theta_2')
df.plot('Unnamed: 0','theta_3', ax=ax[1][0], title = 'Theta_3')
df.plot('Unnamed: 0','theta_4', ax=ax[1][1], title = 'Theta_4')
df.plot('Unnamed: 0','theta_5', ax=ax[2][0], title = 'Theta_5')
df.plot('Unnamed: 0','theta_6', ax=ax[2][1], title = 'Theta_6')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.1, hspace=1.5)
```

```python
#Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot.
h, ax = plt.subplots(3,2, figsize=(10,15))
df.hist('theta_1', bins = 50, ax=ax[0][0])
df.hist('theta_2', bins = 50, ax=ax[0][1])
df.hist('theta_3', bins = 50, ax=ax[1][0])
df.hist('theta_4', bins = 50, ax=ax[1][1])
df.hist('theta_5', bins = 50, ax=ax[2][0])
df.hist('theta_6', bins = 50, ax=ax[2][1])
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
```

```python
#Plot KDE-s (Kernel Denisty Estimators) for all columns.
df.plot.kde()
```
